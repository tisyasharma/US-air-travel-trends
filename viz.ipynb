```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('cleaned_air_traffic.csv')
df.head()
```

```python

# Sort chronologically
df = df.sort_values('date')

# Create a base chart
base = alt.Chart(df).encode(
    x=alt.X('date:T', title='Date'),
    y=alt.Y('total_passengers:Q', title='Number of Passengers')
)

# Passenger trend line
line = base.mark_line(color='#1f77b4').encode(
    tooltip=['date:T', 'total_passengers:Q']
)

# Key global events
events = pd.DataFrame([
    {'event': '2008 Recession', 'date': '2008-09-15'},
    {'event': 'COVID-19 Pandemic', 'date': '2020-03-11'}
])
events['date'] = pd.to_datetime(events['date'])

event_rules = alt.Chart(events).mark_rule(color='red', strokeDash=[4, 2]).encode(
    x='date:T',
    tooltip=['event']
)

event_labels = alt.Chart(events).mark_text(
    align='left', dx=5, dy=-70, color='red', fontSize=11
).encode(
    x='date:T',
    text='event'
)

# Combine all layers
chart = (line + event_rules + event_labels).properties(
    title='Airline Passenger Trends Over Time',
    width=750,
    height=400
).interactive()

chart
```

```python
# Parse date column
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
elif 'year' in df.columns and 'month' in df.columns:
    df['date'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'].astype(str) + '-01')


# Extract month name for grouping
df['month'] = df['date'].dt.month_name()

# Create the boxplot
chart = (
    alt.Chart(df)
    .mark_boxplot(color='steelblue')
    .encode(
        x=alt.X('month:N', sort=['January','February','March','April','May','June',
                                 'July','August','September','October','November','December'],
                title='Month'),
        y=alt.Y('total_passengers:Q', title='Number of Passengers'),
        tooltip=['month', 'total_passengers']
    )
    .properties(
        title='Distribution of Airline Passengers by Month',
        width=700,
        height=400
    )
)

chart
```

```python
import altair as alt

data_2025 = pd.read_csv("data/flight_data_2025.csv")
# remove rows that contain 0.0 in any column
mask = (data_2025 == 0.0).any(axis=1)
data_2025 = data_2025.loc[~mask].reset_index(drop=True)
data_2025.describe()

alt.Chart(data_2025).mark_bar(color='teal').encode(
    x=alt.X('PASSENGERS:Q', bin=alt.Bin(maxbins=50), title='Number of Passengers'),
    y=alt.Y('count()', title='Frequency'),
    tooltip=['count()']
).properties(
    title='Histogram of Passengers (2025 Data)',
    width=600,
    height=350
)
```
